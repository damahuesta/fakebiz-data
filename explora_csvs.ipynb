# Exploración de ficheros CSV generados

Este notebook carga y muestra los datos sintéticos generados en la carpeta `./data/out/`.

```python
# Importar librerías necesarias
import pandas as pd
import os
```

## Listar archivos CSV disponibles

```python
csv_dir = './data/out/'
csv_files = [f for f in os.listdir(csv_dir) if f.endswith('.csv')]
print('Archivos CSV encontrados:')
for f in csv_files:
    print('-', f)
```

## Cargar y mostrar los primeros registros de cada CSV

```python
for f in csv_files:
    print(f'\n--- {f} ---')
    df = pd.read_csv(os.path.join(csv_dir, f))
    display(df.head())
```
